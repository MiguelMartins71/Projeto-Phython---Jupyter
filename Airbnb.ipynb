```python
import pandas as pd

```

```python
airbnb_df = pd.read_excel("airbnb.xlsx")
display(airbnb_df)
```

```python
# display(airbnb_df.loc[1])
airbnb_df.rename(columns={'host_is_superhost':'Superhost', 'cancellation_policy':'Política de Cancelamento'}, inplace=True)
display(airbnb_df)
# Caso Usuário Seja Superhost 
# display(airbnb_df.loc[airbnb_df['Superhost'] == 't'])

#Excluir Colunas di dificil entendimento para o Usuário
# airbnb_df = airbnb_df.drop("longitude", axis=1)

airbnb_df[(airbnb_df.Superhost == "t") & (airbnb_df.property_type == "Apartment")].head()
```

```python
#Especificação De Tabela ( Poderá ser Ajustada de acordo com o Filtro) Média de Quartos para cada Tipo de Propieade e Média com Tipo de Quarto
airbnb_df[["property_type", "accommodates","room_type"]].groupby(["property_type","room_type"]).mean()
```

```python
#Calcullo de média de banheiro e Quartos de Dormir com Propiedade e Tipo do Quarto
airbnb_df[["property_type", "accommodates","room_type","bathrooms","bedrooms"]].groupby(["property_type","room_type","bathrooms","bedrooms"]).mean()
```
